```
spark = SparkSession.builder.appName('Creation Dataframe PySpark').getOrCreate()
header = ['MARKET','SERIES','SYMBOL','SECURITY','PREV_CL_PR','OPEN_PRICE','HIGH_PRICE','LOW_PRICE','CLOSE_PRICE','NET_TRDVAL','NET_TRDQTY','TRADES','HI_52_WK','LO_52_WK']
df = spark.read.options(header = False,inferSchema = True).csv("/FileStore/tables/sharemarket-2.csv")
df.show(5)
```

```
df = df.toDF(*header)
```

```
df = df.drop('_c11')
```

```
df.show(5)
```

```
#Query to display the number of series present in the data
spark.sql("select count(series) from data_table").show()
```

Display the series present in the data.

```
df.createOrReplaceTempView("data_table")
spark.sql("select * from data_table limit 5" ).show()
```

```
%sql select * from data_table
```

```
#Find the sump of all the prices in the each series.

spark.sql("select sum(open_price+high_price+low_price+close_price) from data_table").show()
```

```
#Display security,series with highest net trade value
spark.sql("SELECT series, security, NET_TRDVAL FROM data_table WHERE (security, NET_TRDVAL) IN (SELECT security, MAX(NET_TRDVAL) FROM data_table GROUP BY security)").show()
```

```
#.Display the series whose sum of all prices greater than the net trade value.
result5 = spark.sql("select SERIES, sum(round(PREV_CL_PR + OPEN_PRICE + HIGH_PRICE + LOW_PRICE + CLOSE_PRICE)) as Total_price, sum(NET_TRDVAL) as sum_trade_value from Data_table group by series Having Total_price > sum_trade_value")
result5.show()
```

```
#.Display the series with highest net trade quantity.
result6 = spark.sql('SELECT series, net_trdqty from data_table Where net_trdqty = (Select max(net_trdqty) from data_table)')
result6.show()
```

```
#Display the highest and lowest open price
result7 = spark.sql('Select max(open_price), min(open_price) from data_table')
result7.show()
```

```
#Query to display the series which have trades more than 80
result8 = spark.sql('Select series,trades from data_table where trades > 80')
result8.show()
```

```
#Display the difference between the net trade value net trade quantity for each series.(
result9 = spark.sql('Select series, sum(round(net_trdval - net_trdqty)) as diff_net_trd_and_qty from data_table Group By Series')
result9.show()
```
